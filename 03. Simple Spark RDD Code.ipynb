### Intitalize the Spark Session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

### Create RDD using a list

```python
data = [("apple", 10), ("banana", 5), ("apple", 15), ("orange", 8),
        ("banana", 7), ("apple", 12), ("orange", 6)]
rdd = spark.sparkContext.parallelize(data)
```

```python
rdd
```

### To get number of partitions of the RDD

```python
rdd.getNumPartitions()
```

# RDD Transformations
Here only the transformation lineage is defined. The actual execution doesn't start until the RDD Actions is called

* To apply transformation logic to each element of the RDD

```python
# Convert to (product, (price, 1)) for counting
mapped_rdd = rdd.map(lambda x: (x[0], (x[1], 1)))
```

* To filter out values of the RDD



```python
# Keep only fruits with price > 6
filtered_rdd = mapped_rdd.filter(lambda x: x[1][0] > 6)
```

* To apply aggrigation on the RDD

```python
# Sum prices and count occurrences
reduced_rdd = filtered_rdd.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
```

# RDD Actions
Here actual execution occurs based on the defined lineage

- This runs the RDD lineage and prints the output

```python
reduced_rdd.collect()
```

- This extracts the RDD output to a file

```python
output_path = "/content/output_data/"
reduced_rdd.saveAsTextFile(output_path)
```
