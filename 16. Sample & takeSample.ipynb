```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

```python
data = range(1, 101)  # Numbers 1 to 100
rdd = spark.sparkContext.parallelize(data)
```

# sample() - Transformation
sample(withReplacement, fraction, seed) -> Returns another RDD
- **withReplacement:** True/False (elements can be sampled multiple times but not guaranteed)
- **fraction:** 0.0 to 1.0 (approximate fraction of data)
- **seed:** int (random id/key for reproducibility, which ensures the same sample on each run)

```python
# Elements not repeated/duplicated
rdd.sample(withReplacement=False, fraction=0.1, seed=42).collect()
```

```python
# Elements are repeated
rdd.sample(withReplacement=True, fraction=0.1).collect()
```

```python
# This time elements are not repeated
rdd.sample(withReplacement=True, fraction=0.1).collect()
```

```python
# Excat same sample output as before for seed 42
rdd.sample(withReplacement=False, fraction=0.1, seed=42).collect()
```

```python
# Reused same elements from seed 42 as fraction is reduced
rdd.sample(withReplacement=False, fraction=0.07, seed=42).collect()
```

```python
# Reused few elements from seed 42 and also added additional new elements as fraction is increased
rdd.sample(withReplacement=False, fraction=0.15, seed=42).collect()
```

```python
# Not reused the sample from previous seed 42 beacuse withReplacement parameter is changed
rdd.sample(withReplacement=True, fraction=0.1, seed=42).collect()
```

# takeSample() - Action
Return a fixed-size sampled subset of this RDD.

takeSample(withReplacement, num, seed) -> Returns a list
- **withReplacement:** True/False (elements can be sampled multiple times but not guaranteed)
- **num:** exact number of elements to return
- **seed:** int (random id/key for reproducibility, which ensures the same sample on each run)


```python
# Fixed number of elements with replacement
rdd.takeSample(withReplacement=True, num=20)
```

```python
# Fixed number of elements without replacement
rdd.takeSample(withReplacement=False, num=20)
```
