```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

## Column Object
- `Column` is a generic object that represents a single column in a DataFrame.
- It's part of `pyspark.sql` module

```python
data = [('Alice', 25, 50000), ('Bob', 30, 60000), ('Charlie', 35, 75000), ('David', 28, 55000)]
df = spark.createDataFrame(data, ['name', 'age', 'salary'])
df.show()
```

## Accessing column object

```python
# Using column attribute

df.name
```

```python
# Using column reference

df['name']
```

```python
# Using col() method

from pyspark.sql.functions import col

col('name')
```

## Selecting a column

```python
# Using column attribute

df.select(df.name).show()
```

```python
# Using column reference

df.select(df['name']).show()
```

```python
# Using col() method

df.select(col('name')).show()
```

## Giving column alias

```python
df.name.alias('employee_name')
```

```python
df.select(df.name.alias('employee_name')).show()
```

## Ordering column

```python
# Ascending
df.orderBy(df.name.asc()).show()
```

```python
# Descending
df.orderBy(df.name.desc()).show()
```

```python
# Other options
df.orderBy(df.name.asc_nulls_first()).show()
df.orderBy(df.name.asc_nulls_last()).show()

df.orderBy(df.name.desc_nulls_first()).show()
df.orderBy(df.name.desc_nulls_last()).show()
```

## Cast column

```python
df.select(df.salary.cast('double')).show()
```

## Between method

```python
df.where(df.salary.between(55000, 75000)).show()
```

## Other Key Methods for Column Objects:
- Arithmetic: +, -, *, /
- Comparison: ==, !=, <, >, <=, >=
- Logical: &, |, ~ (and, or, not)
- String: .contains(), .startswith(), .endswith()
- Null handling: .isNull(), .isNotNull(), .fillna()
- Aggregation: .sum(), .avg(), .count(), .min(), .max()
