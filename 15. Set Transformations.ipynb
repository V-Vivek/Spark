```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

## Set oprations
Set operations are useful for combining and comparing datasets.

- union() - Returns the union of two RDDs (allows duplicates)
- intersection() - Returns common elements between two RDDs
- subtract() - Returns elements in first RDD but not in second
- distinct() - Returns distinct elements from an RDD
- cartesian() - Returns Cartesian product of two RDDs

```python
rdd1 = spark.sparkContext.parallelize([1, 2, 3, 4, 5])
rdd2 = spark.sparkContext.parallelize([4, 5, 6, 7, 8])
```

## union()
- Combines two RDDs (allows duplicates)

```python
union_rdd = rdd1.union(rdd2)
union_rdd.collect()
```

## intersection()
- Common elements between RDDs
- Output will not contain any duplicates even if input has the duplicates

```python
intersection_rdd = rdd1.intersection(rdd2)
intersection_rdd.collect()
```

## subtract()
- Elements in first RDD but not in second


```python
subtract_rdd = rdd1.subtract(rdd2)
subtract_rdd.collect()
```

## distinct()
- Remove duplicates


```python
rdd_with_duplicates = spark.sparkContext.parallelize([1, 2, 2, 3, 3, 3, 4])
distinct_rdd = rdd_with_duplicates.distinct()
distinct_rdd.collect()
```

## cartesian()
- Cartesian product of two RDDs


```python
rdd1 = spark.sparkContext.parallelize([1, 2])
rdd2 = spark.sparkContext.parallelize(['a', 'b'])
cartesian_rdd = rdd1.cartesian(rdd2)
cartesian_rdd.collect()
```
