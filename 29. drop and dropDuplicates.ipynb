````python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
````

````python
# Sample data
data = [
    (1, "John", 28, 50000),
    (2, "Jane", 32, 60000),
    (3, "Mike", 25, 45000)
]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "name", "age", "salary"])
df.show()
````

## drop()

- The `drop()` method removes one or more columns from a DataFrame.
- It returns a new DataFrame without the specified columns.

Syntax
```python
df.drop("column_name")
df.drop("col1", "col2", ...)
df.drop(col("column_name"))
```

````python
# Drop a single column
df.drop("age").show()
````

````python
# Drop multiple columns
df.drop("age", "salary").show()
````

````python
# Drop using col()
from pyspark.sql.functions import col
df.drop(col("age")).show()
````

````python
# Drop columns conditionally
columns_to_drop = ["age", "salary"]
df.drop(*columns_to_drop).show()
````

## dropDuplicates()

- The `dropDuplicates()` method removes duplicate rows from a DataFrame based on all columns or specific columns.

Syntax
```python
df.dropDuplicates()  # Remove duplicates based on all columns
df.dropDuplicates(["col1", "col2"])  # Remove duplicates based on specific columns
```

````python
# Sample data with duplicates
data_with_dupes = [
    (1, "John", 28),
    (2, "Jane", 32),
    (1, "John", 28),
    (3, "Mike", 25),
    (2, "Jane", 34)
]

df_dupes = spark.createDataFrame(data_with_dupes, ["id", "name", "age"])
df_dupes.show()
````

````python
# Remove all duplicate rows
df_dupes.dropDuplicates().show()
````

````python
# Remove duplicates based on specific columns
df_dupes.dropDuplicates(["name"]).show()
````

````python
# Remove duplicates based on multiple columns
df_dupes.dropDuplicates(["id", "name"]).show()
````

## Key Differences


| Feature | drop() | dropDuplicates() |
|---------|--------|------------------|
| Purpose | Remove columns | Remove duplicate rows |
| Parameters | Column name(s) | Column name(s) or none (all columns) |
| Affects | Column structure | Row count |
| Alternative method | select() with exclusion | distinct() |
| Use Case | Schema simplification | Data deduplication |
