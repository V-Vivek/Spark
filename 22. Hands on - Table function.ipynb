```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

## table() function

The spark.table() function is used to read a table from the Hive metastore or a registered temporary view and return it as a DataFrame.

### Syntax:
spark.table(tableName)

### Parameters:
- tableName: Name of the table/view to read (string)

### Returns:
DataFrame

### Use Cases:
- Read tables from Hive metastore
- Access temporary views created with createOrReplaceTempView()
- Access global temporary views
- Query Delta tables, Iceberg tables, etc. registered in the metastore

```python
data = [(1, "John", 28, "Engineering"), (2, "Sarah", 32, "Marketing"), (3, "Mike", 45, "Finance"), (4, "Emma", 29, "Engineering")]
columns = ["id", "name", "age", "department"]

df = spark.createDataFrame(data, columns)

# Register as temporary view
df.createOrReplaceTempView("employees")
```

```python
employees_df = spark.table("employees")
employees_df.show()
```
