````python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
````

````python
# Sample data
data = [
    (1, "John", 28, 50000),
    (2, "Jane", 32, 60000),
    (3, "Mike", 25, 45000)
]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "name", "age", "salary"])
df.show()
````

# select()

The `select()` method is used to select specific columns from a DataFrame. It can accept column names as strings or Column objects.

### Syntax
```python
df.select(col1, col2, ...)
```

### Select specific columns

````python
df.select("name", "age").show()
````

````python
# Different ways to select a column
df.select('name', df.name, df.name.alias('employee_name')).show()
````

### Select using col()

````python
from pyspark.sql.functions import col
df.select(col("name"), col("salary")).show()
````

### Select with transformations

````python
from pyspark.sql.functions import upper
df.select("name", upper(col("name")).alias("upper_name")).show()
````

## selectExpr()

The `selectExpr()` method allows you to write SQL-like expressions directly as strings, making it more concise for complex transformations.

### Syntax
```python
df.selectExpr("expression1", "expression2", ...)
```

### Select with SQL expressions

````python
df.selectExpr("name", "age", "salary * 1.1 AS increased_salary").show()
````

### Multiple operations

````python
df.selectExpr(
    "name",
    "age + 5 AS age_plus_five",
    "salary / 12 AS monthly_salary",
    "UPPER(name) AS upper_name"
).show()
````

### Aggregations

````python
df.selectExpr("COUNT(*) AS total_count", "AVG(salary) AS avg_salary").show()
````

## Key Differences


| Feature | select() | selectExpr() |
|---------|----------|--------------|
| Flexibility | Requires functions import | Direct SQL expressions |
| Readability | More verbose | More concise for complex logic |
| Support for SQL functions | Yes, but requires import | Yes, directly in strings |
| Use Case | Simple column selection and transformations | Complex expressions and aggregations |
