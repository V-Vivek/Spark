```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

```python
data = [
    (1, "John Doe", 30, "IT"),
    (2, "Jane Smith", 25, "HR"),
    (3, "Bob Johnson", 35, "Finance")
]
columns = ["id", "name", "age", "department"]
df = spark.createDataFrame(data, columns)
```

### 1. CSV Writing

```python
# Basic CSV write
df.write.csv("output/csv_basic")
```

```python
# CSV with commonly used parameters
df.write \
    .mode("overwrite") \
    .option("header", "true") \
    .option("sep", ",") \
    .option("quote", '"') \
    .option("escape", "\\") \
    .option("compression", "gzip") \
    .csv("output/csv_advanced")
```

```python
# CSV with partition
df.write \
    .mode("overwrite") \
    .partitionBy("department") \
    .option("header", "true") \
    .csv("output/csv_partitioned")
```

### 2. Text Writing

```python
# Text format requires single column DataFrame
text_df = df.select("name")
```

```python
# Basic text write
text_df.write.text("output/text_basic")
```

```python
# Text with commonly used parameters
text_df.write \
    .mode("overwrite") \
    .option("compression", "gzip") \
    .text("output/text_advanced")
```

### 3. Parquet Writing

```python
# Basic Parquet write
df.write.parquet("output/parquet_basic")
```

```python
# Parquet with commonly used parameters
df.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet("output/parquet_advanced")
```

```python
# Parquet with partition
df.write \
    .mode("overwrite") \
    .partitionBy("department") \
    .option("compression", "gzip") \
    .parquet("output/parquet_partitioned")
```

```python
# Parquet with bucketing
df.write \
    .mode("overwrite") \
    .bucketBy(10, "id") \
    .sortBy("age") \
    .saveAsTable("employee_bucketed")
```

### 4. JSON Writing

```python
# Basic JSON write
df.write.json("output/json_basic")
```

```python
# JSON with commonly used parameters
df.write \
    .mode("overwrite") \
    .option("compression", "gzip") \
    .option("dateFormat", "yyyy-MM-dd") \
    .option("timestampFormat", "yyyy-MM-dd HH:mm:ss") \
    .json("output/json_advanced")
```

```python
# JSON with partition
df.write \
    .mode("overwrite") \
    .partitionBy("department") \
    .json("output/json_partitioned")
```

### Common Write Modes
- "overwrite": Overwrite existing data
- "append": Append to existing data
- "ignore": Ignore if data already exists
- "error" or "errorifexists": Throw error if data exists (default)

# Common Compression Options
- CSV/Text: "none", "gzip", "bzip2", "lz4", "snappy", "deflate"
- Parquet: "none", "snappy", "gzip", "lzo", "brotli", "lz4", "zstd"
- JSON: "none", "gzip", "bzip2", "lz4", "snappy", "deflate"

### Additional Useful Parameters

```python
# Coalesce to reduce number of output files
df.coalesce(1).write.csv("output/csv_single_file")
```

```python
# Repartition for better parallelism
df.repartition(4).write.parquet("output/parquet_repartitioned")
```

```python
# Save with specific format method
df.write.format("csv").option("header", "true").save("output/csv_format_method")
```
