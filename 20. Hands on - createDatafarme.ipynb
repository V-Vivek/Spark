```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

# Create DF using list



```python
data_list = [('Robert', 35), ('James', 25)]
```

### Without specifying schema

```python
df = spark.createDataFrame(data = data_list)
df.show()
```

```python
df.printSchema()
```

### With schema - Without datatype

```python
df = spark.createDataFrame(data = data_list, schema = ('Name', 'Age'))
df.show()
```

```python
df.printSchema()
```

### With schema - With datatype

```python
df = spark.createDataFrame(data = data_list, schema = ('Name string, Age int'))
df.show()
```

```python
df.printSchema()
```

# Create DF using list of dict

```python
data_dict = [{'name': 'Robert', 'age': 35}, {'name': 'James','age': 25}]
```

### Without schema

```python
df = spark.createDataFrame(data = data_dict)
df.show()
```

```python
df.printSchema()
```

### With schema and datatype

```python
df = spark.createDataFrame(data = data_dict, schema = ('name string, age int'))
df.show()
```

```python
df.printSchema()
```

# Create DF using RDD

```python
rdd = spark.sparkContext.parallelize(data_list)
rdd.collect()
```

### Without schema

```python
df = spark.createDataFrame(data = rdd)
df.show()
```

```python
df.printSchema()
```

### With schema and datatype

```python
df = spark.createDataFrame(data = rdd, schema = ('Name string, Age int'))
df.show()
```

```python
df.printSchema()
```

# Create DF using Row object

```python
from pyspark.sql import Row
```

```python
# Sample code to understand row object
row = Row(name='Robert', age=35)
print(row)
print(row.name)
print(row.age)
```

```python
df = spark.createDataFrame(data = [Row(name='Robert', age=35), Row(name='James', age=25)])
df.show()
```

```python
df.printSchema()
```
