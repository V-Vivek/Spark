```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

```python
# Sample data
data = [('James', 'Sales', 'NY', 9000, 34),
        ('Alicia', 'Sales', 'NY', 8600, 56),
        ('Robert', 'Sales', 'CA', 8100, 30),
        ('John', 'Sales', 'AZ', 8600, 31),
        ('Ross', 'Sales', 'AZ', 8100, 33),
        ('Kathy', 'Sales', 'AZ', 1000, 39),
        ('Lisa', 'Finance', 'CA', 9000, 24),
        ('Deja', 'Finance', 'CA', 9900, 40),
        ('Sugie', 'Finance', 'NY', 8300, 36),
        ('Ram', 'Finance', 'NY', 7900, 53),
        ('Satya', 'Finance', 'AZ', 8200, 53),
        ('Kyle', 'Marketing', 'CA', 8000, 25),
        ('Reid', 'Marketing', 'NY', 9100, 50)]
df = spark.createDataFrame(data, ['Name', 'Department', 'State', 'Salary', 'Age'])
df.show()
```

### monotonically_increasing_id()
- Generates monotonically increasing 64 bit integers which are unique
- Can be used to generate UUID for each row

```python
from pyspark.sql.functions import monotonically_increasing_id
df.withColumn('key', monotonically_increasing_id()).show()
```

### lit()
- Creates a column of literal value

```python
from pyspark.sql.functions import lit
df.withColumn('Country', lit('USA')).show()
```

```python
# To add literal string between 2 column values

from pyspark.sql.functions import lit, concat
df.withColumn('Branch', concat('Department', lit('-'), 'State')).show()
```

### expr()
- Takes SQL Expression as string argument, executes the SQL and returns a DF column

```python
from pyspark.sql.functions import expr
df.withColumn('name_len', expr('LENGTH(Name)')).show()
```

### spark_partition_id()
- Generates a column with partition IDs

```python
from pyspark.sql.functions import spark_partition_id
df.withColumn('partition_id', spark_partition_id()).show()
```

### rand() and randn()
- rand(): Generates a column with random samples
uniformly distributed in (0.0, 1.0) 📊

- randn(): Generates a column with random samples
from the standard normal distribution. (🔔 Bell curve)

```python
from pyspark.sql.functions import rand, randn
df.withColumn('rand', rand()).withColumn('randn', randn()).show()
```
