````python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
````

````python
# Sample data
data = [
    (1, "John", 28, 50000),
    (2, "Jane", 32, 60000),
    (3, "Mike", 25, 45000)
]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "name", "age", "salary"])
df.show()
````

## withColumn()

- The `withColumn()` method adds a new column or replaces an existing column in a DataFrame.
- It returns a new DataFrame with the specified column added or modified.

Syntax
```python
df.withColumn("column_name", expression)
```

### Add a new column


````python
# With a constant value
from pyspark.sql.functions import lit
df.withColumn("country", lit("USA")).show()
````

````python
# Add a calculated column
from pyspark.sql.functions import col
df.withColumn("annual_bonus", col("salary") * 0.1).show()
````

````python
# Replace an existing column
from pyspark.sql.functions import col
df.withColumn("salary", col("salary") * 1.05).show()
````

## withColumnRenamed()

- The `withColumnRenamed()` method renames an existing column in a DataFrame.
- It returns a new DataFrame with the renamed column.

Syntax
```python
df.withColumnRenamed("existing_name", "new_name")
```

````python
from pyspark.sql.functions import upper
df.select("name", upper(col("name")).alias("upper_name")).show()
````

## selectExpr()

The `selectExpr()` method allows you to write SQL-like expressions directly as strings, making it more concise for complex transformations.

### Syntax
```python
df.selectExpr("expression1", "expression2", ...)
```

### Rename a single column

````python
df.withColumnRenamed("name", "employee_name").show()
````

### Rename multiple columns by chaining

````python
df.withColumnRenamed("name", "employee_name") \
  .withColumnRenamed("age", "employee_age") \
  .withColumnRenamed("salary", "annual_salary") \
  .show()
````

## Key Differences


| Feature | withColumn() | withColumnRenamed() |
|---------|--------------|---------------------|
| Purpose | Add or modify columns | Rename existing columns |
| Parameters | Column name and expression | Old name and new name |
| Creates new column | Yes | No (only renames) |
| Modifies existing column | Yes | No (preserves data) |
| Use Case | Data transformation and addition | Column naming and standardization |
