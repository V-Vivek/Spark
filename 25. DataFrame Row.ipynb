```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

## Row Object
- `Row` is a generic object that represents a single record in a DataFrame.
- It's an ordered collection of fields that can be accessed by index or by field name.
- It's part of `pyspark.sql` module

```python
from pyspark.sql import Row
```

```python
# Sample code to understand row object
row = Row(name='Robert', age=35)
print(f'Row object -> {row}')
print(f'Name -> {row.name}')
print(f'Age -> {row.age}')
```

## Method 1: Create DataFrame using named Row object
- Here we need to pass the key-value pair each time while creating the row object

```python
data = [
    Row(name = 'Alice', age = 35),
    Row(name = 'Bob', age = 25),
    Row(name = 'Chris', age = 40)
]

df = spark.createDataFrame(data)
df.show()
df.printSchema()
```

## Method 2: Create DataFrame using template of Row object
- Once the template is defined we just need to pass the values while creating the row object, key is not required

```python
# Define the template
Person = Row('name', 'age')

data = [
    Person('Alice', 35),
    Person('Bob', 25),
    Person('Chris', 40)
]

df = spark.createDataFrame(data)
df.show()
df.printSchema()
```
