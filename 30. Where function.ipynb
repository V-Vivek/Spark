```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[*]').getOrCreate()
```

```python
ord = spark.read.csv('/content/drive/MyDrive/Spark_Practice/RetailDB_SalesData/Orders/part-00000', schema = ('order_id int, order_date timestamp, order_customer_id int, order_status string'))
ord.printSchema()
ord.show(5)
```

### Where function

```python
# AND condition
ord.where((ord.order_id > 10) & (ord.order_id < 20)).show()
```

```python
# OR condition
ord.where((ord.order_status == 'COMPLETE') | (ord.order_status == 'CLOSED')).show(5)
```

```python
# IN condition
ord.where((ord.order_status.isin('CLOSED', 'COMPLETE'))).show(5)
```

```python
# SQL Style syntax
ord.where("order_status IN ('CLOSED', 'COMPLETE') AND (order_id BETWEEN 10 AND 20)").show()
```
